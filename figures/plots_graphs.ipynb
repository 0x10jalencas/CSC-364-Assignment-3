```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
line_color = "#f72516"

for loss in ["1", "10", "50"]:
    df = pd.read_csv(f"cwnd_log_{loss}.csv")

    plt.figure(figsize=(8, 5))
    plt.plot(df["rtt"], df["cwnd"], color=line_color, linewidth=2)

    plt.xlabel("Time (RTTs)")
    plt.ylabel("Congestion Window (cwnd)")
    plt.title(f"Congestion Window vs RTT ({loss}% Packet Loss)")

    plt.grid(True, alpha=0.3)
    plt.tight_layout()

    plt.savefig(f"cwnd_loss_{loss}.png", dpi=300)
    plt.show()
```

```python
line_color = "#f72516"

for loss in ["1", "10", "50"]:
    df = pd.read_csv(f"retransmission_log_{loss}.csv")

    plt.figure(figsize=(8, 5))
    plt.plot(df["rtt"], df["retransmissions"], color=line_color, linewidth=2)

    plt.xlabel("Time (RTTs)")
    plt.ylabel("Cumulative Retransmissions")
    plt.title(f"Retransmissions vs RTT ({loss}% Packet Loss)")

    plt.grid(True, alpha=0.3)
    plt.tight_layout()

    plt.savefig(f"retransmissions_loss_{loss}.png", dpi=300)
    plt.show()
```
